```
import pandas as pd
```

```
df=pd.read_csv('/Height-Weight Data.csv')
```

```
df
```

```
x=df[['Height']]
```

```
x
```

```
y=df[['Weight']]
```

```
y
```

```
import matplotlib.pyplot as plt
```

```
plt.scatter(df['Height'],df['Weight'],color='blue')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('Height-Wight')
```

```
from sklearn.model_selection import train_test_split
```

```
xtrain,xtest,ytrain,ytest=train_test_split(x,y,train_size=.75)
```

```
xtrain
```

```
xtrain.shape
```

```
xtest
```

```
xtest.shape
```

```
from sklearn.linear_model import LinearRegression
```

```
lr=LinearRegression()
```

```
lr.fit(xtrain,ytrain)
```

```
a=lr.intercept_
a
```

```
b=lr.coef_
b
```

```
lr.score(xtrain,ytrain)
```

```
lr.score(xtest,ytest)
```

```
pred_y=lr.predict(xtest)
```

```
pred_y
```

```
ytest
```

```
df[['New Weight']]=lr.predict(x)
```

```
df
```

```
plt.scatter(df['Height'],df['New Weight'],color='blue')
plt.xlabel('Height')
plt.ylabel('New Weight')
plt.title('Height-Wight')
```
